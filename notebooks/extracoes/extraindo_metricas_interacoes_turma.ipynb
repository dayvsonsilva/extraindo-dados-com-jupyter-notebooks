* Quantidade de teste: é contabilizado a quantidade de teste que o usuário realizou durante a janela de tempo espeficicada;

* Quantidade de submissões corretas: armazena a quatidade de teste submissões corretas que cada usuário fez dentro da janela de tempo;

* Quantidade de submissões incorretas: armazena a quatidade de teste submissões incorretas que cada usuário fez dentro da janela de tempo;

* Media de erros: é soma de todos os erros que cada usuário cometeu ao tentar solucionar o exercício, tanto nos testes quanto nas submissões divido pela quantidade de questões;

* Exercícios feitos: Quantidade de exercícios que o usuário conseguiu finalizar com sucesso;

* Tempo ativo: caso ele passe mais de 60s sem demonstrar qualquer atividade esse intervalo de tempo é desconsiderado, ou seja, não é contabilizado como tempo ativo;

* Quantidade de login: é contabilizado quantos logins o usuário fez durante o intervalo de tempo considerado;

* Média de tempo das submissões: é a média das médias de tempo de cada questão, para a primeira média usa-se a quantidade de submissões para uma questão e a soma dos intervalos entre cada submissão. Para segunda média usa-se a soma das médias de cada questão dividida pela quantidade de questões;


* Média de tempo dos testes: é similar a anterior, o diferencial está na contabilização dos testes e não das submissões;

* Média da quantidade de linhas: é soma da quantidade de linhas de cada exercício dividio pela quantidade de exercícios;

```
def count_correct_submission(start_date, end_date, path):

    str2date = datetime.strptime
    start_date = str2date(start_date, '%Y-%m-%d %H:%M:%S')
    end_date = str2date(end_date, '%Y-%m-%d %H:%M:%S')
    count_t = 0

    with open(path, 'r') as execution_file:
        is_date_in = False
        for line in execution_file:
        
            # verifica se é uma submissão válida dentro do intervalo
            if re.search(r'== SUBMITION \(.+\)', line):
                date = re.search(r'\((.+)\)', line).group(1)
                date = str2date(date, '%Y-%m-%d %H:%M:%S')

                if start_date <= date and date <= end_date:
                    # print(date)
                    is_date_in = True


            # verifica se a questão foi finalizada
            if is_date_in and re.match(r"100\%", line):
                # print(line)
                count_t += 1
                is_date_in = False

    return count_t

count_correct_submission('2016-07-01 19:14:51', '2016-07-06 19:15:41', '/content/codebench/2016-1/110/users/1215/executions/105_1100.log')
```

```
def count_incorret_submission(start_date, end_date, path):

    str2date = datetime.strptime
    start_date = str2date(start_date, '%Y-%m-%d %H:%M:%S')
    end_date = str2date(end_date, '%Y-%m-%d %H:%M:%S')
    count_t = 0

    with open(path, 'r') as execution_file:
        is_date_in = False
        is_submition = False
        for line in execution_file:
        
            # verifica se é uma submissão válida dentro do intervalo
            if re.search(r'== SUBMITION \(.+\)', line):
                date = re.search(r'\((.+)\)', line).group(1)
                date = str2date(date, '%Y-%m-%d %H:%M:%S')

                if start_date <= date and date <= end_date:
                    # print(line)
                    is_date_in = True

            # verifica se a questão foi finalizada
            if is_date_in and (re.match(r"-- GRADE:", line) != None):
                # print(line)
                is_submition = True
            
            # verifica se a sumissão está correta
            n = re.search(r"(\d+)\%", line)
            if is_submition and (n != None):
                if n.group(1) != '100':
                    # print(line)
                    count_t += 1
                    is_date_in = False
                    is_submition = False

    return count_t

count_incorret_submission('2016-07-01 19:14:51', '2016-07-06 19:15:41', '/content/codebench/2016-1/110/users/1215/executions/105_1100.log')
```

```
def count_error_submission(start_date, end_date, path):

    str2date = datetime.strptime
    start_date = str2date(start_date, '%Y-%m-%d %H:%M:%S')
    end_date = str2date(end_date, '%Y-%m-%d %H:%M:%S')
    count_t = 0

    with open(path, 'r') as execution_file:
        is_date_in = False
        for line in execution_file:
        
            # verifica se é uma submissão válida dentro do intervalo
            if re.search(r'== SUBMITION \(.+\)', line):
                date = re.search(r'\((.+)\)', line).group(1)
                date = str2date(date, '%Y-%m-%d %H:%M:%S')

                if start_date <= date and date <= end_date:
                    is_date_in = True

            # verifica se a questão foi finalizada
            if is_date_in and (re.match(r"-- ERROR:", line) != None):
                is_date_in = False
                count_t += 1

    return count_t

count_error_submission('2016-07-01 19:14:51', '2016-07-06 19:15:41', '/content/codebench/2016-1/110/users/1215/executions/105_1100.log')
```

```

def count_tests(start_date, end_date, path):
    '''Extract amount done tests.
    
    This function counts the number of times it finds the word test and 
    checks whether the test performed is within the indicated range.

    Keyword arguments:
    star_date -- first date, begin of interval
    end_date -- second date, end of interval
    path -- file path o read
    '''
    str2date = datetime.strptime
    start_date = str2date(start_date, '%Y-%m-%d %H:%M:%S')
    end_date = str2date(end_date, '%Y-%m-%d %H:%M:%S')
    count_t = 0

    with open(path, 'r') as execution_file:
        for line in execution_file:
            if re.search(r'== TEST \(.+\)', line):
                date = re.search(r'\((.+)\)', line).group(1)
                date = str2date(date, '%Y-%m-%d %H:%M:%S')
                if start_date <= date and date <= end_date:
                    count_t +=1
   
    return count_t

count_tests('2016-07-01 19:14:51', '2016-07-06 19:15:41', '/content/codebench/2016-1/110/users/1215/executions/105_1100.log')
```

## Extraindo a quatidade logins

```
def amount_logins(start_date, end_date, path):
    
    str2date = datetime.strptime
    start_date = str2date(start_date, '%Y-%m-%d %H:%M:%S')
    end_date = str2date(end_date, '%Y-%m-%d %H:%M:%S')
    count_t = 0

    with open(path, 'r') as login_file:
        for line in login_file:
            if re.search(r'login.', line):
                dt = re.search(r'(.+)\#(.+)', line).group(1)
                dt = str2date(dt, '%Y-%m-%d %H:%M:%S')
                if start_date <= dt and dt <= end_date:
                    count_t += 1
    return count_t

amount_logins('2016-06-07 00:00:00', '2016-07-19 00:00:00','/content/codebench/2016-1/102/users/1278/logins.log')
```

## Extraindo a quantidade de tempo ativo dentro da plataforma

[Explicações sobre os eventos que acontecem na interação do usuário com a IDE](http://codemirror.net/doc/manual.html#events)


**** O que foi considerado aqui para extrair essa métrica, foi se o usuário passou mais de 60 segundos entre um evento e outro então é um intervalo válido, depois esses intervalo é somado junto aos demais

```
def get_active_time(start_date, end_date, path):
    
    str2date = datetime.strptime
    start_date = str2date(start_date, '%Y-%m-%d %H:%M:%S')
    end_date = str2date(end_date, '%Y-%m-%d %H:%M:%S')
    accumulator = 0

    is_first = True

    with open(path, 'r') as codemirror_file:
        for line in codemirror_file:
            # dt = re.search('(.+)\.\d+\#', line)
            # 2018-4-14 15:09:28.358#viewportChange#0(
            dt = re.search('\d{4}\-\d{1,2}\-\d{1,2}\s\d{1,2}\:\d{1,2}\:\d{1,2}', line)
            if dt is not None:
                dt = dt.group(0)
                if dt == '2016-8-99 15:10:06': dt = '2016-8-9 15:10:06'
                dt = str2date(dt, '%Y-%m-%d %H:%M:%S')
                if start_date <= dt and dt <= end_date:
                    
                    # armazena a primeira data
                    if is_first == True:
                        is_first = False
                        dtf = dt
                    
                    diff_date_time = dt - dtf
                    # print(f'diff={diff_date_time.seconds}')
                    if float(diff_date_time.seconds) <= 180.0:
                        accumulator += float(diff_date_time.seconds)
                        # print(f'accumulator={accumulator}')

                    # o anterior recebe o atual
                    dtf = dt
    return accumulator

# get_active_time('2018-3-6 21:26:01', '2018-3-8 20:38:41', '/content/codebench/2018-1/187/users/3221/codemirror/1019_1329.log')
get_active_time('2016-05-31 00:00:00', '2016-09-13 12:45:00', '/content/codebench/2016-1/102/users/1315/codemirror/208_1280.log')
```

## Média dos intervalos de tempo das submissões

```
def average_interval_submission(start_date, end_date, path):

    str2date = datetime.strptime
    start_date = str2date(start_date, '%Y-%m-%d %H:%M:%S')
    end_date = str2date(end_date, '%Y-%m-%d %H:%M:%S')
    count_t = 0
    is_first = True
    accumulator = 0

    with open(path, 'r') as execution_file:
        for line in execution_file:
        
            # verifica se é uma submissão válida dentro do intervalo
            if re.search(r'== SUBMITION \(.+\)', line):
                date = re.search(r'\((.+)\)', line).group(1)
                date = str2date(date, '%Y-%m-%d %H:%M:%S')

                if start_date <= date and date <= end_date:
                    # conta a quantidade de intervalos
                    count_t += 1

                    # armazena a primeira data
                    if is_first == True:
                        is_first = False
                        prev_date = date

                    # acumula os intervalos entre as submissões
                    date_diff = date - prev_date
                    accumulator += float(date_diff.seconds)
                    prev_date = date

    if count_t <= 1:
        average = 0.0
    else:
        average = accumulator/(count_t-1)
    
    # print(round(average, 3))
    return round(average, 3)


average_interval_submission('2016-07-01 19:14:51', '2016-07-06 19:15:41', '/content/codebench/2016-1/110/users/1215/executions/105_1100.log')
```

## Extraindo a quantidade linhas de código válidas

***** As linhas de código válidas são linhas diferentes de commentários e linhas em branco

```
def count_valid_lines(path):
    count_lines = 0

    with open(path, 'r') as file_code:
        for line in list(file_code.readlines()):
            if not '#' in line and '\n' != line:
                count_lines += 1
                
    return count_lines


count_valid_lines('/content/codebench/2018-1/185/users/2146/codes/1039_835.py')
```

```
def dataframe_users(path_user, semester_id, class_id, sdate, edate):

    metrics = []
    
    features = [
        'semester_id',
        'class_id',
        'user_id',

        'number_logins',

        'average_active_time',
        'average_exam_sub_comp',
        'average_exam_sub_inc',
        'average_exam_sub_err',
        'average_exam_test',
        'average_exam_test_err',
        'average_exam_code_line',
        'average_exam_sub_average_interval',

        'average_homework_sub_comp',
        'average_homework_sub_inc',
        'average_homework_sub_err',
        'average_homework_test',
        'average_homework_test_err',
        'average_homework_code_line',
        'average_homework_test_average'
    ]

    for user_id in os.listdir(path_user):        
        # quantidade de invervalos ativos
        count_active_times = 0
        # quantidade de assessments do tipo homework
        count_amount_homework = 0
        # quantidade de assessments do tipo exam
        count_amount_exam = 0
        # quantidade de logins do usuário
        number_logins = amount_logins(sdate, edate, os.path.join(path_user, user_id, 'logins.log'))

        # tempo ativo na ide
        path_codemirror = os.path.join(path_user, user_id, 'codemirror/')
        amount_active_time = 0
        for codemirror_file in os.listdir(path_codemirror):
            assessment_id = re.sub('\_.+', '', codemirror_file)
            if is_assessment_valid(assessment_id):
                # print(path_codemirror, codemirror_file)
                amount_active_time += get_active_time(sdate, edate, os.path.join(path_codemirror, codemirror_file))
                count_active_times += 1

        path_executions = os.path.join(path_user, user_id, 'executions/')

        # extraindo os dados de submissão e testes
        exam_submissions_complete = 0
        exam_submissions_incomplete = 0
        exam_submissions_error = 0
        exam_submissions_average_interval = 0
        exam_testes_amount = 0
        exam_testes_amount_errors = 0
        exam_testes_average_interval = 0
        exam_amount_code_line = 0

        homework_submissions_complete = 0
        homework_submissions_incomplete = 0
        homework_submissions_error = 0
        homework_submissions_average_interval = 0
        homework_testes_amount = 0
        homework_testes_amount_errors = 0
        homework_testes_average_interval = 0
        homework_amount_code_line = 0

        for execution_file in os.listdir(path_executions):
            path_execution_file = os.path.join(path_executions, execution_file)
            print(path_execution_file)
            assessment_id = re.sub('\_\d+.log', '', execution_file)
            # print(path_execution_file)
        
            # verificando se é uma questão válida
            if is_assessment_valid(assessment_id):

                # verificando se a questão é exam
                if verify_type_assessment(assessment_id) == 'exam':

                    # extraindo_dados_submisssion
                    exam_submissions_complete += count_correct_submission(sdate, edate, path_execution_file)
                    exam_submissions_incomplete += count_incorret_submission(sdate, edate, path_execution_file)
                    exam_submissions_error += count_error_submission(sdate, edate, path_execution_file)
                    exam_submissions_average_interval += average_interval_submission(sdate, edate, path_execution_file)

                    # dados do teste
                    exam_testes_amount += count_tests(sdate, edate, path_execution_file)
                    exam_testes_amount_errors += count_error_tests(sdate, edate, path_execution_file)
                    exam_testes_average_interval += average_interval_tests(sdate, edate, path_execution_file)

                    # linhas de código de exam
                    exam_amount_code_line += count_valid_lines(path_execution_file)

                    # quantidades de assessments exam
                    count_amount_exam += 1
                
                # verificando se a questão é homework
                elif verify_type_assessment(assessment_id) == 'homework':
                    
                    # extraindo_dados_submisssion
                    homework_submissions_complete += count_correct_submission(sdate, edate, path_execution_file)
                    homework_submissions_incomplete += count_incorret_submission(sdate, edate, path_execution_file)
                    homework_submissions_error += count_error_submission(sdate, edate, path_execution_file)
                    homework_submissions_average_interval += average_interval_submission(sdate, edate, path_execution_file)

                    # dados do teste
                    homework_testes_amount += count_tests(sdate, edate, path_execution_file)
                    homework_testes_amount_errors += count_error_tests(sdate, edate, path_execution_file)
                    homework_testes_average_interval += average_interval_tests(sdate, edate, path_execution_file)

                    # quantidade de linhas de código de homework
                    homework_amount_code_line += count_valid_lines(path_execution_file)

                    # quantidade de assessments homework
                    count_amount_homework += 1
        
        # tirando a média das métricas dentro da janela de tempo
        if count_active_times == 0: count_active_times = 1
        if count_amount_exam == 0: count_amount_exam = 1
        if count_amount_homework == 0: count_amount_homework = 1

        average_active_time = round(amount_active_time / count_active_times, 3)
        average_exam_sub_comp = round(exam_submissions_complete / count_amount_exam, 3)
        average_exam_sub_inc = round(exam_submissions_incomplete / count_amount_exam, 3)
        average_exam_sub_err = round(exam_submissions_error / count_amount_exam, 3)
        average_exam_test = round(exam_testes_amount / count_amount_exam, 3)
        average_exam_test_err = round(exam_testes_amount_errors / count_amount_exam, 3)
        average_exam_code_line = round(exam_amount_code_line / count_amount_exam, 3)
        average_exam_sub_average_interval = round(exam_submissions_average_interval / count_amount_exam, 3)

        average_homework_sub_comp =  round(homework_submissions_complete / count_amount_homework, 3)
        average_homework_sub_inc = round(homework_submissions_incomplete / count_amount_homework, 3)
        average_homework_sub_err = round(homework_submissions_error / count_amount_homework, 3)
        average_homework_test = round(homework_testes_amount / count_amount_homework, 3)
        average_homework_test_err = round(homework_testes_amount_errors / count_amount_homework, 3)
        average_homework_code_line = round(homework_amount_code_line / count_amount_homework, 3)
        average_homework_test_average_interval =round(homework_testes_average_interval / count_amount_homework, 3)

        values = [
            semester_id,
            class_id,
            user_id,
            
            number_logins,

            average_active_time,

            average_exam_sub_comp,
            average_exam_sub_inc,
            average_exam_sub_err,
            average_exam_test,
            average_exam_test_err,
            average_exam_code_line,
            average_exam_sub_average_interval,

            average_homework_sub_comp,
            average_homework_sub_inc,
            average_homework_sub_err,
            average_homework_test,
            average_homework_test_err,
            average_homework_code_line,
            average_homework_test_average_interval
        ]
        metrics.append( dict(zip(features, values)) )
        
    df_metrics = pd.DataFrame.from_dict(metrics, orient='columns')
    return df_metrics

def data_metrics_users(sdate, edate, semester_id='', class_id=''):
    df_metrics = pd.DataFrame()
    if semester_id == '' and class_id == '':
        for semester_id in os.listdir(root):
            path_class = os.path.join(root, semester_id)
            for class_id  in os.listdir(path_class):
                path_user = os.path.join(path_class, class_id, 'users/')
                df = dataframe_users(path_user,  semester_id, class_id, sdate, edate)
                df_metrics.append(df)
    else:
            path_user = f'codebench/{semester_id}/{class_id}/users/'
            df_metrics = dataframe_users(path_user, semester_id, class_id, sdate, edate)
    return df_metrics
            
 
# data_metrics_users('2016-01-01 00:00:00', '2020-01-01 00:00:00', '2016-1', '102')
```
