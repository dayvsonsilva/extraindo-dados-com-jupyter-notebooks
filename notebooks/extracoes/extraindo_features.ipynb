## Verificando se a avalição é uma avaliação monitorada

> Requisito 1: é uma avaliação que está no dataframe de avalições válidas.

> Requisito 2: o id do assessment é menor do que o id do exam desejado

```
def is_assessment_valid(start_date, end_date, class_id, assessment_id, end_number):
    str2date = datetime.strptime
    start_date = str2date(start_date, '%Y-%m-%d %H:%M:%S')
    end_date = str2date(end_date, '%Y-%m-%d %H:%M:%S')

    r1 = len(df_assessment.loc[(df_assessment['assessment_id'] == int(assessment_id))]) == 1
    if r1:
        id = df_number_exam.loc[(df_number_exam['class_id'] == int(class_id)) & (df_number_exam['exam_number'] == int(end_number))].iloc[0, 2]
        r2 = int(assessment_id) <= id
        if r2:
            return True
    return False   
```

## Funções de extração de dados dos arquivos

### Contando a quantidade de submissões corretas de uma questão

> As vezes um usuário envia mais de uma submissão correta mais de uma vez

```
def count_correct_submission(start_date, end_date, path):

    str2date = datetime.strptime
    start_date = str2date(start_date, '%Y-%m-%d %H:%M:%S')
    end_date = str2date(end_date, '%Y-%m-%d %H:%M:%S')
    count_t = 0

    with open(path, 'r') as execution_file:
        is_date_in = False
        for line in execution_file:
        
            # verifica se é uma submissão válida dentro do intervalo
            if re.search(r'== SUBMITION \(.+\)', line):
                date = re.search(r'\((.+)\)', line).group(1)
                date = str2date(date, '%Y-%m-%d %H:%M:%S')

                if start_date <= date and date <= end_date:
                    # print(date)
                    is_date_in = True


            # verifica se a questão foi finalizada
            if is_date_in and re.match(r"100\%", line):
                # print(line)
                count_t += 1
                is_date_in = False

    return count_t
```

### Contando a quantidaded submissões que foram compiladas, mas não produziram a saída esperada de cada questão do usuário

```
def count_incorret_submission(start_date, end_date, path):

    str2date = datetime.strptime
    start_date = str2date(start_date, '%Y-%m-%d %H:%M:%S')
    end_date = str2date(end_date, '%Y-%m-%d %H:%M:%S')
    count_t = 0

    with open(path, 'r') as execution_file:
        is_date_in = False
        is_submition = False
        for line in execution_file:
        
            # verifica se é uma submissão válida dentro do intervalo
            if re.search(r'== SUBMITION \(.+\)', line):
                date = re.search(r'\((.+)\)', line).group(1)
                date = str2date(date, '%Y-%m-%d %H:%M:%S')

                if start_date <= date and date <= end_date:
                    # print(line)
                    is_date_in = True

            # verifica se a questão foi finalizada
            if is_date_in and (re.match(r"-- GRADE:", line) != None):
                # print(line)
                is_submition = True
            
            # verifica se a sumissão está correta
            n = re.search(r"(\d+)\%", line)
            if is_submition and (n != None):
                if n.group(1) != '100':
                    # print(line)
                    count_t += 1
                    is_date_in = False
                    is_submition = False

    return count_t
```

### Contando a quantidade de submissões que não foram compiladas, que produziram um erro na hora da compilação

```
def count_error_submission(start_date, end_date, path):

    str2date = datetime.strptime
    start_date = str2date(start_date, '%Y-%m-%d %H:%M:%S')
    end_date = str2date(end_date, '%Y-%m-%d %H:%M:%S')
    count_t = 0

    with open(path, 'r') as execution_file:
        is_date_in = False
        for line in execution_file:
        
            # verifica se é uma submissão válida dentro do intervalo
            if re.search(r'== SUBMITION \(.+\)', line):
                date = re.search(r'\((.+)\)', line).group(1)
                date = str2date(date, '%Y-%m-%d %H:%M:%S')

                if start_date <= date and date <= end_date:
                    is_date_in = True

            # verifica se a questão foi finalizada
            if is_date_in and (re.match(r"-- ERROR:", line) != None):
                is_date_in = False
                count_t += 1

    return count_t
```

### Contando a quantidade de testes de cada questão produzida pelo usuário

```
def count_tests(start_date, end_date, path):
    str2date = datetime.strptime
    start_date = str2date(start_date, '%Y-%m-%d %H:%M:%S')
    end_date = str2date(end_date, '%Y-%m-%d %H:%M:%S')
    count_t = 0

    with open(path, 'r') as execution_file:
        for line in execution_file:
            if re.search(r'== TEST \(.+\)', line):
                date = re.search(r'\((.+)\)', line).group(1)
                date = str2date(date, '%Y-%m-%d %H:%M:%S')
                if start_date <= date and date <= end_date:
                    count_t +=1
    return count_t
```

### Contando a quantidade de erros na compilação de testes de cada questão do usuário

```
def count_error_tests(start_date, end_date, path):

    str2date = datetime.strptime
    start_date = str2date(start_date, '%Y-%m-%d %H:%M:%S')
    end_date = str2date(end_date, '%Y-%m-%d %H:%M:%S')
    count_t = 0

    with open(path, 'r') as execution_file:
        is_date_in = False
        for line in execution_file:
            
            if re.search(r'== TEST \(.+\)', line):
                date = re.search(r'\((.+)\)', line).group(1)
                date = str2date(date, '%Y-%m-%d %H:%M:%S')
                if start_date <= date and date <= end_date:
                    # print(line)
                    is_date_in = True


            if is_date_in and (re.search(r'-- ERROR:', line) != None):
                # print(line)
                count_t += 1
                is_date_in = False
    return count_t
```

### Extraindo a quatidade logins

```
def amount_logins(start_date, end_date, path):
    
    str2date = datetime.strptime
    start_date = str2date(start_date, '%Y-%m-%d %H:%M:%S')
    end_date = str2date(end_date, '%Y-%m-%d %H:%M:%S')
    count_t = 0

    with open(path, 'r') as login_file:
        for line in login_file:
            if re.search(r'login.', line):
                dt = re.search(r'(.+)\#(.+)', line).group(1)
                dt = str2date(dt, '%Y-%m-%d %H:%M:%S')
                if start_date <= dt and dt <= end_date:
                    count_t += 1
    return count_t
```

### Extraindo a quantidade de tempo ativo dentro da plataforma

[Explicações sobre os eventos que acontecem na interação do usuário com a IDE](http://codemirror.net/doc/manual.html#events)


**** O que foi considerado aqui para extrair essa métrica, foi se o usuário passou mais de 60 segundos entre um evento e outro então é um intervalo válido, depois esses intervalo é somado junto aos demais

```
def get_active_time(start_date, end_date, path):
    
    str2date = datetime.strptime
    start_date = str2date(start_date, '%Y-%m-%d %H:%M:%S')
    end_date = str2date(end_date, '%Y-%m-%d %H:%M:%S')
    accumulator = 0

    is_first = True

    with open(path, 'r') as codemirror_file:
        for line in codemirror_file:
            dt = re.search('\d{4}\-\d{1,2}\-\d{1,2}\s\d{1,2}\:\d{1,2}\:\d{1,2}', line)
            if dt is not None:
                dt = dt.group(0)
                if dt == '2016-8-99 15:10:06': dt = '2016-8-9 15:10:06'
                dt = str2date(dt, '%Y-%m-%d %H:%M:%S')
                if start_date <= dt and dt <= end_date:
                    
                    if is_first == True:
                        is_first = False
                        dtf = dt
                    
                    diff_date_time = dt - dtf
                    # print(f'diff={diff_date_time.seconds}')
                    if float(diff_date_time.seconds) <= 180.0:
                        accumulator += float(diff_date_time.seconds)
                    dtf = dt
    return accumulator
```

### Extraindo a média dos intervalos de tempo entre cada uma das submissões por questão

```
def average_interval_submission(start_date, end_date, path):

    str2date = datetime.strptime
    start_date = str2date(start_date, '%Y-%m-%d %H:%M:%S')
    end_date = str2date(end_date, '%Y-%m-%d %H:%M:%S')
    count_t = 0
    is_first = True
    accumulator = 0

    with open(path, 'r') as execution_file:
        for line in execution_file:
        
            # verifica se é uma submissão válida dentro do intervalo
            if re.search(r'== SUBMITION \(.+\)', line):
                date = re.search(r'\((.+)\)', line).group(1)
                date = str2date(date, '%Y-%m-%d %H:%M:%S')

                if start_date <= date and date <= end_date:
                    # conta a quantidade de intervalos
                    count_t += 1

                    # armazena a primeira data
                    if is_first == True:
                        is_first = False
                        prev_date = date

                    # acumula os intervalos entre as submissões
                    date_diff = date - prev_date
                    accumulator += float(date_diff.seconds)
                    prev_date = date

    if count_t <= 1:
        average = 0.0
    else:
        average = accumulator/(count_t-1)
    return round(average, 4)
```

### Média dos invervalos de tempo entre cada teste feito em cada exercício

```
def average_interval_tests(start_date, end_date, path):

    str2date = datetime.strptime
    start_date = str2date(start_date, '%Y-%m-%d %H:%M:%S')
    end_date = str2date(end_date, '%Y-%m-%d %H:%M:%S')
    count_t = 0
    is_first = True
    accumulator = 0

    with open(path, 'r') as execution_file:
        is_date_in = False
        for line in execution_file:
            
            if re.search(r'== TEST \(.+\)', line):
                date = re.search(r'\((.+)\)', line).group(1)
                date = str2date(date, '%Y-%m-%d %H:%M:%S')
                if start_date <= date and date <= end_date:
                    # conta a quantidade de intervalos
                    count_t += 1

                    # armazena a primeira data
                    if is_first == True:
                        is_first = False
                        prev_date = date

                    # acumula os intervalos entre as submissões
                    date_diff = date - prev_date
                    accumulator += float(date_diff.seconds)
                    prev_date = date

    if count_t <= 1:
        average = 0.0
    else:
        average = accumulator/(count_t-1)
    return round(average, 3)
```

### Extraindo a quantidade linhas de código válidas

**As linhas de código válidas são linhas diferentes de commentários e linhas em branco.**

```
def count_valid_lines(path):
    count_lines = 0
    with open(path, 'r') as file_code:
        for line in list(file_code.readlines()):
            if not '#' in line and '\n' != line:
                count_lines += 1
    return count_lines
```

#### Função para extração de ativação de tempo ativo 

```
def average_active_time(sdate, edate, class_id, path_codemirror, end_number):
    at = []
    for codemirror_file in os.listdir(path_codemirror):
        assessment_id = re.sub('\_.+', '', codemirror_file)
        if is_assessment_valid(sdate, edate, class_id,  assessment_id, end_number):
            p = os.path.join(path_codemirror, codemirror_file)
            at.append(get_active_time(sdate, edate, p))
    return round(np.mean(at), 4)
```

```


features = [
    'semester_id',
    'class_id',
    'user_id',

    'number_logins',

    'average_active_time',
    
    'average_exam_sub_comp',
    'average_exam_sub_inc',
    'average_exam_sub_err',
    'average_exam_sub_mean_time',

    'average_exam_test',
    'average_exam_test_err',
    'average_exam_test_mean_time',
    
    'average_exam_code_line',

    'average_homework_sub_comp',
    'average_homework_sub_inc',
    'average_homework_sub_err',
    'average_homework_sub_mean_time',

    'average_homework_test',
    'average_homework_test_err',
    'average_homework_test_mean_time',

    'average_homework_code_line',

    'sum_exam_sub_comp',
    'sum_exam_sub_inc',
    'sum_exam_sub_err',
    'sum_exam_sub_mean_time',

    'sum_exam_test',
    'sum_exam_test_err',
    'sum_exam_test_mean_time',
    
    'sum_exam_code_line',

    'sum_homework_sub_comp',
    'sum_homework_sub_inc',
    'sum_homework_sub_err',
    'sum_homework_sub_mean_time',

    'sum_homework_test',
    'sum_homework_test_err',
    'sum_homework_test_mean_time',

    'sum_homework_code_line',    
]

class Generate_dataframe_class:
    def __init__(self):
        self.sum_data_user_ = []
        self.mean_data_user_ = []
        self.data_user = []
        self.sdate = ''
        self.edate = ''
        self.root = 'codebench/'
        self.count = 0
        self.size_ = 21
        self.register = {}
        self.list_df = []


    def sum_data_user(self):
        self.sum_data_user_ = [[] for i in range(self.size_)]
        for i in range(len(self.data_user)):
            if i >= 5:
                self.sum_data_user_[i] = round(np.sum(self.data_user[i]), 4)
            else:
                self.sum_data_user_[i] = self.data_user[i][0]
    
    def mean_data_user(self):
        self.mean_data_user_ = [[] for i in range(self.size_)]
        for i in range(len(self.data_user)):
            if i >= 5:
                self.mean_data_user_[i] = round(np.mean(self.data_user[i]), 4)
            else:
                self.mean_data_user_[i] = self.data_user[i][0]


    def extract_exam(self, path_execution_file, path_code):
        # submisssion
        self.data_user[5].append(count_correct_submission(self.sdate, self.edate, path_execution_file))
        self.data_user[6].append(count_incorret_submission(self.sdate, self.edate, path_execution_file))
        self.data_user[7].append(count_error_submission(self.sdate, self.edate, path_execution_file))
        self.data_user[8].append(average_interval_submission(self.sdate, self.edate, path_execution_file))
        #testes
        self.data_user[9].append(count_tests(self.sdate, self.edate, path_execution_file))
        self.data_user[10].append(count_error_tests(self.sdate, self.edate, path_execution_file))
        self.data_user[11].append(average_interval_tests(self.sdate, self.edate, path_execution_file))
        # linhas de código de exam
        self.data_user[12].append(count_valid_lines(path_code))
    
    def extract_homework(self, path_execution_file, path_code):
        # submisssion
        self.data_user[13].append(count_correct_submission(self.sdate, self.edate, path_execution_file))
        self.data_user[14].append(count_incorret_submission(self.sdate, self.edate, path_execution_file))
        self.data_user[15].append(count_error_submission(self.sdate, self.edate, path_execution_file))
        self.data_user[16].append(average_interval_submission(self.sdate, self.edate, path_execution_file))
        # test
        self.data_user[17].append(count_tests(self.sdate, self.edate, path_execution_file))
        self.data_user[18].append(count_error_tests(self.sdate, self.edate, path_execution_file))
        self.data_user[19].append(average_interval_tests(self.sdate, self.edate, path_execution_file))
        # quantidade de linhas de código de homework
        self.data_user[20].append(count_valid_lines(path_code))


    def extract_data_execution_file(self, path_executions, path_user, user_id, class_id, end_number):
        for execution_file in os.listdir(path_executions):
            path_execution_file = os.path.join(path_executions, execution_file)
            assessment_id       = re.sub('\_\d+.log', '', execution_file)
            code_file           = execution_file.replace('.log', '.py')
            path_code           = os.path.join(path_user, user_id, 'codes/', code_file)
            # verificando se é uma questão válida
            # requisito 1: é uma avaliação monitorada
            # requisito 2: a avaliação está range do intervalo
            if is_assessment_valid(self.sdate, self.edate, class_id, assessment_id, end_number):
                # verificando se a questão é exam
                if verify_type_assessment(assessment_id) == 'exam':
                    self.extract_exam(path_execution_file, path_code)
                # verificando se a questão é homework
                elif verify_type_assessment(assessment_id) == 'homework':
                    self.extract_homework(path_execution_file, path_code)


    def create_user_register(self, path_user, semester_id, class_id, user_id, end_number):
        self.data_user = [[]  for i in range(self.size_)]
        self.data_user[0].append(semester_id)        
        self.data_user[1].append(class_id)
        self.data_user[2].append(user_id)

        ### number_logins
        p = os.path.join(path_user, user_id, 'logins.log')
        self.data_user[3].append(amount_logins(self.sdate, self.edate, p))

        ### average_active_time
        p = os.path.join(path_user, user_id, 'codemirror/')
        self.data_user[4].append(average_active_time(self.sdate, self.edate, class_id, p, end_number))

        path_executions = os.path.join(path_user, user_id, 'executions/')
        self.extract_data_execution_file(path_executions, path_user, user_id, class_id, end_number)
        
        self.mean_data_user()
        self.sum_data_user()

        v = self.mean_data_user_ + self.sum_data_user_[5:]
        self.register = dict(zip(features, v))
        #print(self.register)

    def dataframe_users(self, path_user, semester_id, class_id, end_number):
        list_register = []
        count = 0
        for user_id in os.listdir(path_user):
            # print('  ', user_id)
            self.create_user_register(path_user, semester_id, class_id, user_id, end_number)
            list_register.append(self.register)
            count += 1
            #if count > 20:
            #print(list_register)
        df = pd.DataFrame(list_register)
        print(df)
        self.list_df.append(df)
    
    def select_dates_end_point(self, class_id, end_point):
        self.sdate = df_first_dates.loc[df_first_dates['class_id'] == int(class_id)].iloc[0, 4]+':00'
        self.edate = df_number_exam.loc[(df_number_exam['class_id'] == int(class_id)) &  (df_number_exam['exam_number'] == end_point)].iloc[0, 4]+':00'
        print(self.sdate, self.edate)
    
    def select_dates_start_end_point(self, class_id, start_point, end_point):
        self.sdate = df_number_exam.loc[(df_number_exam['class_id'] == int(class_id)) & (df_number_exam['exam_number'] == start_point)].iloc[0, 4]+':00'
        self.edate = df_number_exam.loc[(df_number_exam['class_id'] == int(class_id)) & (df_number_exam['exam_number'] == end_point)].iloc[0, 4]+':00'

    def select_function(self, class_id, start_number, end_number):
        if start_number == '':
            self.select_dates_end_point(class_id, end_number)
        else:
            self.select_dates_stcreate_user_registerart_end_point(class_id, start_number, end_number)
    
    def select_all_semesters(self, start_number, end_number):
        df_metrics = pd.DataFrame()
        for semester_id in os.listdir(self.root):
            print(semester_id)
            path_class = os.path.join(self.root, semester_id)
            for class_id  in os.listdir(path_class):
                self.select_function(class_id, start_number, end_number)
                path_user = os.path.join(path_class, class_id, 'users/')
                self.dataframe_users(path_user,  semester_id, class_id, end_number)

    def select_first_semesters(self, start_number, end_number):
        df_metrics = pd.DataFrame()
        count = 0
        for semester_id in os.listdir(self.root):
            if not re.search(r'\-1', semester_id): continue
            print(semester_id)

            path_class = os.path.join(self.root, semester_id)
            for class_id  in os.listdir(path_class):
                #if count == 2: break
                print(' ', class_id)
                self.select_function(class_id, start_number, end_number)
                path_user = os.path.join(path_class, class_id, 'users/')
                self.dataframe_users(path_user,  semester_id, class_id, end_number)
                count += 1

    def select_class(self, semester_id, class_id, start_number, end_number):
        self.select_function(class_id, start_number, end_number)
        path_user = f'codebench/{semester_id}/{class_id}/users/'
        self.dataframe_users(path_user, semester_id, class_id, end_number)   


    def create_dataframe(self, start_number='', end_number='', semester_id='', class_id='', first_semesters=False):
        if first_semesters == True:
            self.select_first_semesters(start_number, end_number)

        elif semester_id == '' and class_id == '':
            self.select_all_semesters(start_number, end_number)

        else:
            self.select_class(semester_id, class_id, start_number, end_number)
            
        df_treino = pd.concat(self.list_df, ignore_index=True)
        return df_treino
```
